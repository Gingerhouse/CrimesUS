```python
import pandas as pd
from pycaret.regression import *
```

```python
df = pd.read_csv('crimedata.csv')
```

```python
s = setup(df, 
          target = 'ViolentCrimesPerPop',
          silent = True,
          categorical_features = ['state', 'LemasGangUnitDeploy'],
          ignore_features = ['countyCode', 'communityCode', 'communityName'],
          ordinal_features = {'LemasGangUnitDeploy' : [0.0, 5.0, 10.0]},
          imputation_type = 'iterative',
          numeric_iterative_imputer = 'et',
          categorical_iterative_imputer = 'et',
          normalize = True,
          remove_outliers = True,
          feature_selection = True,
          remove_multicollinearity = True,
          pca = True,
          pca_components = 10)
```

```python
lr = create_model('lr')
```

```python
lr
```

```python
evaluate_model(lr)
```

```python
predict_model(lr)
```

```python
final_lr = finalize_model(lr)
```

```python
save_model(final_lr, 'lr_model')
```
